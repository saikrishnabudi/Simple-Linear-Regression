```python
import statsmodels.formula.api as smf
```

```python
model1=smf.ols("Salary~np.sqrt(YearsExperience)",data=df).fit()
model1
```

```python
model1.summary()
```

```python
pred1=model1.predict(df.YearsExperience)
pred1
```

```python
model1.resid
```

```python
rmse_1=np.sqrt(np.mean((np.array(df["YearsExperience"])-np.array(pred1))**2))
rmse_1
```

```python
plt.scatter(x=df["YearsExperience"],y=df["Salary"],color="Blue")
plt.plot(df["YearsExperience"],pred1,color="Black")
plt.xlabel("Yearsof Experience")
plt.show()
```

```python
model2=smf.ols("Salary~np.log(YearsExperience)",data=df).fit()
model2
```

```python
model2.summary()
```

```python
pred2=model2.predict(df.YearsExperience)
pred2
```

```python
model2.resid
```

```python
rmse_log=np.sqrt(np.mean((np.array(df["YearsExperience"])-np.array(pred2))**2))
rmse_log
```

```python
pred2.corr(df.Salary)
```

```python
plt.scatter(x=df["YearsExperience"],y=df["Salary"],color="Maroon")
plt.plot(df["YearsExperience"],pred2,color="Blue")
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.show()
```

```python
model3=smf.ols("Salary~np.cbrt(YearsExperience)",data=df).fit()
model3
```

```python
model3.summary()
```

```python
pred3=model3.predict(df.YearsExperience)
pred3
```

```python
model3.resid
```

```python
rmse_log=np.sqrt(np.mean((np.array(df["YearsExperience"])-np.array(pred3))**2))
rmse_log
```

```python
pred3.corr(df.Salary)
```

```python
plt.scatter(x=df["YearsExperience"],y=df["Salary"],color="purple")
plt.plot(df["YearsExperience"],pred3,color="green")
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.show()
```
